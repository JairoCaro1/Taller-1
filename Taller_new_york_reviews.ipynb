```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
review = pd.read_csv('reviews.csv')
```

```python
review.shape
```

```python
review.info()
```

Se Convierte la columna date a tipo to_datetime

```python
review['date'] = pd.to_datetime(review['date']).dt.date
```

```python
review['date'].max()
```

```python
review['date'].min()
```

```python
review.head()
```

Se toma la decision de solamente tener encuenta la reseñas de los dos ultimos años para usar unicamente las informacion mas actualizada

```python
fecha_comparacion = pd.to_datetime('2022-07-05').date()
```

```python
review = review[review['date'] >= fecha_comparacion]
```

```python
review['date'].min()
```

```python
review.head()
```

Se segmenta la tabla para sacar el numero de reseñas por departamento

```python
review_list = review.groupby(['listing_id']).agg(
                            review_count=('reviewer_id', 'count')
                            ).reset_index()
```

```python
review_list
```

Se genera el nuevo dataset de reviews

```python
review_list.to_csv('review_list.csv', index=False,sep=',')
```
